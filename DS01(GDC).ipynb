# GCD


```python
def gcd_sub(a,b):
  while a !=0 and b !=0:
    if a>b:
      a =a-b
    else:
      b = b-a
  return a+b
a = int(input())
b = int(input())
x = gcd_sub(a,b)
print(x)
```

```python
def gcd_mob(a,b):
  while a !=0 and b !=0:
    if a>b:
      a =a%b
    else:
      b = b%a
  return a+b
a = int(input())
b = int(input())
y = gcd_mob(a,b)
print(y)
```

```python
def gcd_rec(a,b):
  while a !=0 and b != 0:
    if a>b:
      return(gcd_rec(a-b,b))
    else: 
      return(gcd_rec(a,b-a))
  return a+b

a = int(input())
b = int(input())
z = gcd_rec(a,b)
print(z)
```

```python
def gcd_sub(a,b):
  while a !=0 and b !=0:
    if a>b:
      a =a-b
    else:
      b = b-a
  return a+b

def gcd_mob(a,b):
  while a !=0 and b !=0:
    if a>b:
      a =a%b
    else:
      b = b%a
  return a+b

def gcd_rec(a,b):
  while a !=0 and b != 0:
    if a>b:
      return(gcd_rec(a-b,b))
    else: 
      return(gcd_rec(a,b-a))
  return a+b

dt = input()

a = int(dt.split()[0])
b = int(dt.split()[1])

x = gcd_sub(a,b)
y = gcd_mob(a,b)
z = gcd_rec(a,b)

print(x,y,z)
```
