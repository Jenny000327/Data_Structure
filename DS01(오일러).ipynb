# 오일러 수 수행 시간 비교

```python
import time, signal

class TimeOutException(Exception):
  pass

def alarm_handler(signum, frame):
    print("Time is up!")
    raise TimeOutException()
 
signal.signal(signal.SIGALRM, alarm_handler)
signal.alarm(60)


def compute_e_ver1(n): # O(n^2)
  result = 1 
  if n < 1:
    return 1
  for i in range(1, n+1):
    temp = 1
    for j in range(1, i+1):
      temp *= j 
    result += 1 / temp
  return result
  

def compute_e_ver2(n): # O(n)
  lst = [1]
  result = 1
  for i in range(1, n+1):
    lst.append(lst[i-1] * i)
    result += 1 / (lst[i-1] * i)
  return result

#__main__
n = int(input())

# 두 함수의 각각 수행시간 출력 
try:
    before = time.process_time()
    result = compute_e_ver1(n)
    after = time.process_time()
    time1 = after-before
    print(time1) 
    print(result)
except TimeOutException as e:
    print(e)

try:
    before2 = time.process_time()
    result2 = compute_e_ver2(n)
    after2 = time.process_time()
    time2 = after2-before2
    print(time2) 
    print(result2)
except TimeOutException as e:
    print(e)
   

```
