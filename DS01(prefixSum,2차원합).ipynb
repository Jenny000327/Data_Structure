# PrefixSum

```python
import time, random
```

```python
def prefixSum1(X, n): 
  S = list(0 for _ in range(n))
  for i in range(n):
    for j in range(i + 1):
      S[i] += X[j]
  return S[n-1]
```

```python
def prefixSum2(X, n):
  S = list(0 for _ in range(n))
  for i in range(n):
    S[i] = S[i-1]+X[i]
  return S[n-1]
```

```python
#main
random.seed() #random 함수 초기화
n = int(input())
X = [random.randint(-999, 999) for _ in range(n)]
print(X)

#prefixSum1 수행시간
before = time.process_time()
prefixSum1(X,n)
after = time.process_time()
time1 = after-before
print('prefixSum1 수행시간: ', time1)
print('prefixSum1 수행결과: ',prefixSum1(X,n))

#prefixSum2 수행시간
before = time.process_time()
prefixSum2(X, n)
after = time.process_time()
time2 =after-before
print('prefixSum2 수행시간: ', time2)
print('prefixSum2 수행결과: ', prefixSum2(X, n))

#수행 시간 비교
if(time1 < time2):
  print('prefixSum1 수행시간이 더 짧습니다.')
elif(time2 < time1):
  print('prefixSum2 수행시간이 더 짧습니다.')
else:
  print('수행시간이 동일합니다.')
```

```python
import time, random

def prefixSum1(X, n):#O(n**2)
	# code for prefixSum1
	S = list(0 for _ in range(n))
	for i in range(n):
		for j in range(i+1):
			S[i] += X[j]
	return S[n-1]


def prefixSum2(X, n): #O(n)
	# code for prefixSum2
	S = list(0 for _ in range(n))
	for i in range(n):
		S[i] = S[i-1] +X[i]
	return S[n-1]
	
random.seed()		# random 함수 초기화
n = int(input()) # n 입력받음
X = [random.randint(-999,999) for _ in range(n)]# 리스트 X를 randint를 호출하여 n개의 랜덤한 숫자로 채움

# prefixSum1 호출
before= time.process_time()
prefixSum1(X,n)
after = time.process_time()
time1 = after-before

# prefixSum2 호출
before = time.process_time()
prefixSum2(X,n)
after = time.process_time()
time2 = after-before

# 두 함수의 수행시간 출력
print('prefixSum1 수행시간: ', time1)
print('prefixSum2 수행시간: ', time2)
```

# 2차원 합 계산

```python
import time, random

#이차원 합
def sum(A,n):
  B= [[0 for _ in range(n)] for _ in range(n)]
  for i in range(n):
    for j in range(n):
      if i == j:
        B[i][j] =A[j]
      elif i < j:
        B[i][j] = B[i][j-1] + A[j]
  return B

n = int(input())
#1차원 배열
A = [random.randint(0,100) for _ in range(n)] 
print(A)

# sum 함수의 수행시간 측정
before = time.process_time()
result = sum(A, n)
after = time.process_time()
time1 = after - before

print(time1)
print(result)
```
