```python
# 1. class Stack 정의 
class Stack:
  def __init__(self):
    self.items = []           # 데이터 저장을 위한 리스트 준비
  
  def push(self, val): 
    self.items.append(val)

  def pop(self): 
    try:
      return self.items.pop()  # pop할 아이템이 없으면 
    except IndexError:         # indexError 발생
      print("Stack is empty")

  def top(self): 
    try:
      return self.items[-1] 
    except IndexError:
      print("Stack is empty")

  def __len__(self):            # len()로 호출하면 stack의 item 수 반환
    return len(self.items)
  
  def isEmpty(self): 
    return self.__len__() == 0
  
# 2. get_token_list 함수 정의 
def get_token_list(expr):
  expr = expr.replace(' ', '')
  return expr

# 3. compute_postfix 함수 정의 
def compute_postfix(token_list):
  sk = Stack()
  for one in token_list:
    # 연산자이면
    if one in '+-/*^':
      n1 = sk.items.pop()
      n2 = sk.items.pop()
      if one == '+':
        sk.push(float(n2) + float(n1))
      elif one == '-':
        sk.push(float(n2) - float(n1))
      elif one == '/':
        sk.push(float(n2) / float(n1))
      elif one == '*':
        sk.push(float(n2) * float(n1))
      elif one == '^':
        sk.push(float(n2) ** float(n1))
    # 피연산자이면 
    else:
      sk.push(one)

  if sk.isEmpty():
    return 'None'
  return sk.top()


# 4. infix_to_postfix 함수 정의 
def infix_to_postfix(token_list):
  opstack = Stack() # 괄호와 연산자 저장 
  outstack = []  # postfix 수식 
  
  token_list = []
  j = 0 
  for i in range(len(token_list)):
    if token_list[i] in '+-/*^()':
      if i != j:
        token_list.append(token_list[j:i])
      token_list.append(token_list[i])
      j = i+1

  if j <= i:
    token_list.append(token_list[j])

  # 연산자의 우선순위 설정 
  prec = {}
  prec['('] = 0 
  prec['+'] = 1
  prec['-'] = 1
  prec['*'] = 2  
  prec['/'] = 2 
  prec['^'] = 3 

  for token in token_list:
    if token == '(': # 우선순위 젤 낮음 : 닫힌괄호나와야 출력되도록하기위함 
      opstack.push(token)
    elif token == ')':
      while not opstack.isEmpty():
        x = opstack.pop()
        if x == '(':
          break
        outstack.append(x)
      else:
          raise ValueError("괄호가 맞지 않습니다.")
    elif token in '+-/*^':
      # opstack에 있는 자신보다 우선순위가 높거나 같은 연산자는
      #  차례로 모두 pop한 후 outstack에 append함 
      while not opstack.isEmpty():
        x = opstack.top()
        if x is not None and prec[x] >= prec[token]: 
          outstack.append(opstack.pop())
        else:
          break
      opstack.push(token)
    else: # 피연산자 
      outstack.append(token)

  while not opstack.isEmpty():
    outstack.append(opstack.pop())
  
  return outstack



# __main__
expr = input()
# expr = '3+ 2* 4/(6- 1)'
# expr = '3.14*2.10/(10.2-21)'
# expr = '6*4.5-18+3.3/3+2^3+(3-11.0*2+0.7)/2'
value = compute_postfix(infix_to_postfix(get_token_list(expr)))
print(value)
```
